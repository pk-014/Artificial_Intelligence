## K Means
Unsupervised Machine Learning Algorithm

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('College_Data',index_col=0)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df2 = df.copy()
df2 = df2.drop('Private',axis=1)
```

```python
df2.head(3)
```

### Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, max_iter=60)
```

```python
predicted_clusters = kmeans.fit_predict(df2)
```

```python
predicted_clusters
```

```python
len(predicted_clusters)
```

### Evaluation

```python
# For evaluation, we need to convert the values of "Private" feature/column in the dataframe 
# from string to numerical value
```

```python
df['Private'].values
```

```python
conv_val = lambda a: 0 if a=="Yes" else 1
```

```python
conv_val
```

```python
df['Cluster'] = [conv_val(item) for item in df['Private'].values]
```

```python
df['Cluster'].values
```

```python
df
```

```python
df['Cluster']
```

```python
# predicted_clusters
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],predicted_clusters))
print(classification_report(df['Cluster'], predicted_clusters))
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(df['Cluster'], predicted_clusters)
print(f"KMeans Accuracy is : {acc}")
```
