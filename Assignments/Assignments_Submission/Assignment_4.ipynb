##### Multi-level Inheritence
 Q1:Make a chain of classes, each one inheriting from the previous one as shown below.
 
 Cake -> CreamCake -> CherryCreamCake -> ChocolateCherryCreamCake

 Methods of Cake: beat_eggs_whites, mix, bake,

 Attributes of CherryCreamCake: cherry
 
 Methods of CherryCreanmCake: add_cherry,
 
 Attributes of ChocolateCherryCreamCake: chocolate,
 
 Methods of Cake: add_chocolate

```python
class cake():
    def __init__(self,floor,egg,sugar,milk,baking_soda,temperature,is_baked):
        self.floor=floor
        self.egg=egg
        self.sugar=sugar
        self.milk=milk
        self.baking_soda=baking_soda
        self.temperature=temperature
        self.is_baked=is_baked
    def beat_eggs_whites(self):
        print("Beat Eggs Whites")
    def mix(self):
        print("Mix the eggs")
    def bake(self):
        print("Bake the Cake")
# Defining CreamCake Class
class creamcake(cake):
    def __init__(self,floor,egg,sugar,milk,baking_soda,temperature,is_baked,cream):
        self.cream=cream
    def add_cream(self):
        print("Add Some Cream")
    def refrigerate(self):
        print("Keep in refrigerator")
# Defining CherryCreamCake Class
class CherryCreamCake(creamcake):
    def __init__(self,floor,egg,sugar,milk,baking_soda,temperature,is_baked,cream,cherry):
        self.cherry=cherry
    def add_cherry(self):
        print("Add some Cherries")
# Defining ChocolateCherryCreamCake Class
class ChocolateCherryCreamCake(CherryCreamCake):
    def __init__(self,floor,egg,sugar,milk,baking_soda,temperature,is_baked,cream,cherry,chocolate):
        self.chocolate=chocolate
    def add_chocolate(self):
        print("Add some chocolate")
```

```python
# Defining Animal Class
class Animal():
    def __init__(self,can_fly,feathers,lives_on,eats):
        self.can_fly=can_fly
        self.feathers=feathers
        self.lives_on=lives_on
        self.eats=eats
    def reproduction(self):
        print("Animals show reproduction")
    def abilities(self):
        print("Animals can walk, run or jump but cannot fly")
# Defining Herbivorous Class
class Herbivorous(Animal):
    def __init__(self,can_fly,feathers,lives_on,eats,teeth,danger_level):
        self.teeth=teeth
        self.danger_level=danger_level
    def what_eat(self):
        print("Herbivorous Eat Plants")
    def hilal(self):
        print("Almost Every Herbivorous is Hilal")
# Defining Cow Class
class Cow(Herbivorous):
    def __init__(self,can_fly,feathers,lives_on,eats,teeth,danger_level,legs,skin):
        self.legs=legs
        self.skin=skin
    def speak(slef):
        print("Cow Speak like 'moo'")
    def products(self):
        print("Cow give us milk, meat and leather etc.")
# Defining SahiwalCow Class
class SahiwalCow(Cow):
    def __init__(self,can_fly,feathers,lives_on,eats,teeth,danger_level,legs,skin,hump):
        self.hump=hump
    def speak(self):
        print("SahiwalCow speak like 'mawmaw'")
    def nature(self):
        print("SahiwalCow are quiet in nature.")
        

# Defining Carnivorous Class
class Carnivorous(Animal):
    def __init__(self,can_fly,feathers,lives_on,eats,can_harm):
        self.can_harm=can_harm
    def danger_level(self):
        print("Some Carnivorous Are dangerous")
    def king(self):
        print("The king of forest is Carnovorous")
# Defining Shark Class
class Shark(Carnivorous):
    def __init__(self,can_fly,feathers,lives_on,eats,can_harm,gills):
        self.gills=gills
    def attack_humans(self):
        print("Sharks attack Humans")
    def eat_human(self):
        print("Sharks Eats Humans(Not as their food) ")
# Defining BlueShark Class
class BlueShark(Shark):
    def __init__(self,can_fly,feathers,lives_on,eats,can_harm,gills,color):
        self.color=color
    def shark_size(self):
        print("BlueShark is twice of an elephant")
    def is_distinct(self):
        print("BlueSharks are on the verge of distinction")
```

##### Multiple Inheritence
Q3:Make three classes as mentioned below and make atleast 2 methods and few data-members of your choice for each class. Also, first make the UML diagram on the page before coding the scenario. Understand the concepts of multiple inheritence and Method-Resolution-Order and how Super() keyword acts to resolve the issue.

Animal -> Carnivorous -> Omnivorous

(Hint: Omnivorous is inheriting from both Herbi & Carnivorous)

```python
# Defining Animal Class
class Animal():
    def __init__(self,name,color,feathers):
        self.name=name
        self.color=color
        self.feathers=feathers
    def cannot_fly(self):
        print("Animals Cannot Fly")
    def species(self):
        print("Animals have many Species")
        
# Defining Herbivorous Class
class Herbivorous(Animal):
    def __init__(self,name,color,feathers):
        super().__init__(name,color,feathers)
    def danger_level(self):
        print("Herbivorous are less dangerous.")
    def useful(self):
        print("Herbivorous are useful.")

# Defining Carnivorous Class
class Carnivorous(Animal):
    def __init__(self,name,color,feathers):
        super().__init__(name,color,feathers)
    def dangerous(self):
        print("They are dangerous")
    def can_attack(self):
        print("They can attack")

# Defining Omnivorous Class
class Omnivorous(Carnivorous,Herbivorous):
    pass

# Creating Object(hedgehog)
hedgehog=Omnivorous("Spike","brown",False)
print(hedgehog.can_attack())
print(hedgehog.useful())
```
